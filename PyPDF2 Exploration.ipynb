```python
# Extract Shares Alloted
def shares_alloted(text_array):
    for text in text_array:
        match_number_alloted = re.search(r'(Number allotted)', text)
        if (match_number_alloted):
            text_match = match_number_alloted
            text_contain = text
#             print(text_match)
            print("Number allotted (extracted): ", text_contain.split(' ')[-1])
            
        
        match_amount_paid = re.search(r'(Amount paid)', text)
        if (match_amount_paid):
            text_match = match_amount_paid
            text_contain = text
#             print(text_match)
            print("Amount paid (extracted): ", text_contain.split(' ')[-1])
            
        
        
        match_amount_unpaid = re.search(r'(Amount unpaid)', text)
        if (match_amount_unpaid):
            text_match = match_amount_unpaid
            text_contain = text
#             print(text_match)
            print("Amount unpaid (extracted): ", text_contain.split(' ')[-1])
            
        
        match_nominal_value = re.search(r'(nominal value)', text, re.IGNORECASE)
        if (match_nominal_value):
            text_match = match_nominal_value
            text_contain = text
#             print(text_match)
            print((" ").join(text_contain.split(' ')[0:-1]) + "(extracted)", (".").join(re.findall(r'\d+',text_contain.split(' ')[-1])))
            
#             break
```

```python
for p in reader.pages:
    text = p.extract_text()
    text_ar = text.split('\n')
#     print("arr, ",text_ar)
    print("========== FULL TEXT ON PAGE ==========")
    print(text)
    print('')    
    
    print ('========= EXTRACTED DATA =========')
    shares_alloted(text_ar)
    print('')
    print('================================')
    print('========== PAGE BREAK ==========')
    print('================================')

    print('\n\n')    
```
